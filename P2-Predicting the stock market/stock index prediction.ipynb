```python
import pandas as pd
from datetime import datetime
from datetime import timedelta

df = pd.read_csv("sphist.csv")
pd.to_datetime(df["Date"])
df.sort_values("Date", inplace=True)
```

```python
df.set_index("Date", inplace=True)
```

```python
df
```

```python
df["day_5_avg"] = 0
df["day_30_avg"] = 0
df["day_365_avg"] = 0
df["day_5_std"] = 0
df["day_365_std"] = 0


df.index = pd.to_datetime(df.index)
for row in df.iterrows():
    #dt = datetime.strptime(row[0],'%Y-%m-%d')
    dt = row[0]
    if dt - datetime(year=1950, month=1, day=3) > timedelta(days=4):
        df.loc[dt,"day_5_avg"] = df.loc[(dt-timedelta(days=6)):(dt-timedelta(days=1)), "Close"].mean()
        df.loc[dt,"day_5_std"] = df.loc[(dt-timedelta(days=6)):(dt-timedelta(days=1)), "Close"].std()
        
    if dt - datetime(year=1950, month=1, day=3) > timedelta(days=29):
        df.loc[dt,"day_30_avg"] = df.loc[(dt-timedelta(days=31)):(dt-timedelta(days=1)), "Close"].mean()
        
    if dt - datetime(year=1950, month=1, day=3) > timedelta(days=364):
        df.loc[dt,"day_365_avg"] = df.loc[(dt-timedelta(days=366)):(dt-timedelta(days=1)), "Close"].mean()
        df.loc[dt,"day_365_std"] = df.loc[(dt-timedelta(days=366)):(dt-timedelta(days=1)), "Close"].std()
        
        
```

```python
df.reset_index(inplace=True)
df = df[df["Date"] > datetime(year=1951, month=1, day=2)]
df.dropna(axis=0, inplace=True)
train = df[df["Date"] < datetime(year=2014, month=1, day=1)]
test = df[df["Date"] > datetime(year=2013, month=12, day=31)]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

lr = LinearRegression()
lr.fit(train[["day_30_avg","day_365_avg","day_365_std"]], train["Close"])
predictions = lr.predict(test[["day_30_avg","day_365_avg","day_365_std"]])
print(mean_absolute_error(test["Close"],predictions))
```
